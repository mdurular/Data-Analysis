```python
# Basic Libraries
import pandas as pd
import numpy as np
import scipy
# Libraries for visualizaton
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
#%matplotlib notebook
```

```python
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset ("tips")
tips.head()
```

```python
tips.describe()
```

```python
tips.day.value_counts()
```

```python
sns.countplot(x = "day", data = tips);
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x='day', data=tips)

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+0.5));
    # matplotlib.pyplot.annotate(text, xy, *args, **kwargs)
    
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x='day', data=tips, hue = "sex")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.15, p.get_height()+0.5));
    # matplotlib.pyplot.annotate(text, xy, *args, **kwargs)
    ax.text(p.get_x()+0.1, p.get_height()-p.get_height()/2,str(round(p.get_height()/tips.day.count(),2)) )     
    # y√ºzdelik ekledik
    
```

# Line Plot

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
sns.lineplot(x = "year", y = "passengers", data=flights);
```

```python
f_sum = flights.groupby(["year", "month"]).sum()
f_sum
```

```python
sns.lineplot(y="passengers", x=f_sum.reset_index().index, data=f_sum);
```

```python
flights_wide = flights.pivot(index = "year", columns = "month", values = "passengers")
flights_wide
```

```python
sns.lineplot(data=flights, x = "year", y = "passengers",  hue = "month");
```

```python
sns.lineplot(data=flights_wide);
```
