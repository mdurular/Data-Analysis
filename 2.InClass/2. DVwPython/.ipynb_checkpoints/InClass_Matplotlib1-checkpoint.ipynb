```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
```

```python
plt.plot(age, salary)
plt.xlabel('Age ')
plt.ylabel('Salary')
plt.title("Salary by Age")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(age,salary, 'r')
ax.set_xlabel("age")
ax.set_ylabel('salary')
ax.set_title("Salary by Age")
plt.show()
```

```python
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
plt.plot(age, salary)
plt.plot(age, salary_2)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");
plt.show()
```

```python
x = np.arange(0,11)
y = x.copy()
z = x*2
t = np.log(x)
```

```python
fig, ax = plt.subplots()

ax.plot(x,y)
ax.plot(x,z)
ax.plot(x,t)

ax.set_xlabel("X")
ax.set_ylabel("Y-Z")
ax.set_title("Title")
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x, z)
plt.plot(x,t)
plt.xlabel("x")
plt.ylabel("y,z,t")
plt.title("x ' e göre y,z,t");
plt.show()
```

```python
fig = plt.figure()      # ax = fig.add_axes([left, bottom, widht, height])
                                          # left, bottom, width, height (range 0 to 1)
ax1 = fig.add_axes([0.0, 0.0, 1, 1])

ax2 = fig.add_axes([1.1, 0.3, 0.4, 0.4])

ax1.plot(age,salary, "r")
ax1.set_xlabel("Age")
ax1.set_ylabel("Salary")
ax1.set_title("Salary by Age")

ax2.plot(age,salary_2, "b");
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary_2")
ax2.set_title("Salary_2 by Age");

plt.show()
```

```python
# functional
plt.subplot(2,1,1)
plt.plot(age, salary, "r")
plt.subplot(2,1,2)
plt.plot(age, salary_2, "b")
plt.show()
```

```python
# object oriented
fig, ax = plt.subplots(2,1)
ax[0].plot(age,salary,"r")
ax[1].plot(age,salary,"b")
plt.show()
```
