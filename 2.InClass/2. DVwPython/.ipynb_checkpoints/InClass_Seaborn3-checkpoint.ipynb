```python
# Basic Libraries
import pandas as pd
import numpy as np
import scipy
# Libraries for visualizaton
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
#%matplotlib notebook
```

```python
print(sns.get_dataset_names())
```

```python
titanic = sns.load_dataset ("titanic")
titanic.head()
```

```python
sns.catplot(x = "pclass", y = "survived", data = titanic, kind = "bar", 
            estimator=np.sum, hue = "sex", col = "alone", row = "embarked");
```

```python
sns.catplot(x = "pclass", y = "survived", data = titanic, kind = "bar", 
            estimator=np.sum, col = "sex");
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.catplot( x= "day", y = "total_bill", data = tips, kind ="box", hue = "smoker");
```

```python
sns.catplot( x= "day", y = "total_bill", data = tips, kind ="violin", hue = "smoker");
```

```python
sns.catplot( x= "day", y = "total_bill", data = tips, kind ="swarm", hue = "smoker");
```

```python
sns.catplot( x= "day", data = tips, kind ="count", hue = "smoker");
```

# Jointplot

```python
sns.jointplot(x='total_bill',y='tip',data=tips);
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind = "hex");
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind = "reg");
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind = "resid");
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind = "kde");
```

# Pairplot

```python
sns.pairplot(tips);
```

```python
sns.pairplot(tips,hue='sex');
```

```python
sns.pairplot(tips,hue='sex',palette= "viridis");
```

```python
sns.pairplot(tips,hue='sex', corner= True, diag_kind="hist");
```

# Grid Plot
# PairGrid

```python
g=sns.PairGrid(tips)
g.map(plt.scatter);
```

```python
g=sns.PairGrid(tips)

g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
#g.map_lower(sns.kdeplot);
```

```python
g=sns.PairGrid(tips)

g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, color ="r");
```

```python
g = sns.PairGrid(tips, hue="sex", palette="viridis")

g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.distplot)
g = g.map_lower(sns.kdeplot)

g = g.add_legend();
```

```python
g = sns.FacetGrid(data = tips, col="time", row="smoker")
g = g.map(plt.hist, "total_bill")
```
