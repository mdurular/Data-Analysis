```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

Data visualization is the act of taking information (data) and placing it into a visual context, 
such as a map or graph. Data visualizations make data easier for the human brain to understand. 
You can more easily detect patterns, trends, and outliers in groups of data.

Good data visualizations should give meaning to your data and clearly communicate what is happening in your analysis. 
Excel has been a go-to data visualization tool for many years. Often, data visualization does not need to be fancy.
As long as your audience understands your work, it is effective data visualization.

**Matplotlib** is a Python 2D plotting library which produces publication quality figures

**Seaborn** is a python data visualization library based on matplotlib. 
It provides a high-level interface for drawing attractive and informative statistical graphics. 
Seems to be accepted as an extension to matplotlib functionality, particularly for statistical visualization.

**Pandas** is an open source python library providing high-performance, easy-to-use data structures and data analysis tools.

**The ggplot** python library evolved out of the ggplot2 R-specific package.

**Bokeh** is different in that it does not depend on matplotlib and is geared toward generating visualizations in the web browser. 
It is meant to make interactive web visualizations.

**Scatter Plot**

Scatter plots show the relationship between two variables(continuous numerical) in the form of dots on the plot. 
In simple terms, the values along a horizontal axis are plotted against a vertical axis.

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
plt.figure(figsize=(8,4),dpi=100)
sns.scatterplot(x="total_bill", y="tip",hue='smoker',data=tips,palette='magma',size='size',alpha=0.7,s=50,style='time')
plt.legend(loc=(1.01,0.5));

# plt.savefig('scatter.jpg')
```

**Pandas Scatterplot**

```python
fig,ax = plt.subplots()
tips.plot(kind='scatter',x='total_bill',y='tip',c='size',s=100,figsize=(12,5),ax=ax)
ax.set_title('Scatter plot of total_bill vs tip') 
ax.set_xlabel('total_bill')
ax.set_ylabel('tip');

# tips.plot.scatter('total_bill','tip',c='size',s=100,figsize=(12,5));
```

**Distribution Plots**

Rugplot---Histogram---KDE Plot

**Rugplot**

```python
sns.rugplot(data=tips, x="total_bill",y="tip", hue="time",height=.15);
```

**Histogram**

Histograms are used for numeric variables. Whereas, a bar plot shows the counts of unique categories,
a histogram shows the number of data with values within a bin. 
The bins divide the values of the variable into equal segments.
The vertical axis of the histogram shows the count of data values within each bin.

```python
sns.histplot(x='tip',data=tips,bins=12,color='blue',edgecolor='black',ls='--',kde=True);
```

```python
sns.set_style('darkgrid')
sns.displot(x='tip',data=tips,bins=12,edgecolor='white',lw=2,rug=True,
                                                         rug_kws={"color": "r", "alpha": 0.2, "linewidth": 2, "height":0.1});
```

**Pandas**

```python
tips.tip.plot.hist(bins=12,grid=True);
```

**KDE Plot**

Kernel density plots are similar in concept to a histogram. 
A kernel density plot displays the values of a smoothed density curve of the data values. 
In other words, the kernel density plot is a smoothed version of a histogram.

```python
sns.kdeplot(x='tip',data=tips,clip=[1,10],bw_adjust=1);
```

 **Bar Plot**
    
Bar plots are used to display the counts of unique values of a categorical variable. 
The height of the bar represents the count for each unique category of the variable.

```python
y=[59219,55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
```

```python
fig, ax = plt.subplots(figsize=(15, 7))
sns.barplot(x=x, y=y,ax=ax ,estimator=np.mean,ci='sd')
ax.set_title("Most Popular Languages")
ax.set_ylabel('Number of People Who Use')
# ax.set_xticks(x)
ax.set_xticklabels(x, rotation=45)
;
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x="day",y="tip",data=tips,hue='smoker' ,estimator=np.mean,ci='sd')
plt.legend(bbox_to_anchor=(1.05,0.5));
```

**Box Plot**

Box plots are another way to visualize the distribution of data values. 
In this respect, box plots are comparable to histograms, but are quite different in presentation.

```python
sns.boxplot(x=tips["total_bill"],orient="h");
```

```python
sns.boxplot(x="day",y="total_bill",data=tips);
```

```python
sns.boxplot(x="day",y="total_bill",hue="smoker",data=tips);
```

**Violin Plot**

A violin plot combines attributes of boxplots and a kernel density estimation plot.
We can say it is a mirrored kde plot

```python
sns.violinplot(x="day",y="total_bill",hue="sex",data=tips);
```

```python
sns.violinplot(x="day",y="total_bill",hue="sex", split=True,data=tips);
```

**Strip Plot**

Draw a scatterplot where one variable is categorical.

```python
sns.stripplot(x="day",y="total_bill", hue="sex",dodge=False, jitter=False, data=tips);
```

```python
sns.stripplot(x="day",y="total_bill", hue="sex" ,dodge=False, jitter=False, data=tips);
plt.legend(loc=3);
```

**Swarm Plot**

Draw a categorical scatterplot with non-overlapping points.
This function is similar to stripplot(), 
but the points are adjusted (only along the categorical axis) so that they dont overlap. 

```python
sns.swarmplot(x="day",y="total_bill",data=tips);
```

```python
sns.swarmplot(x="day",y="total_bill",hue="sex",dodge=False, data=tips);
```

**Boxen Plot**

Similar to boxplot but Instead of quartiles you can see all the quantiles of data

```python
sns.boxenplot(x="day",y="total_bill",data=tips);
```

**Joint Plot**

Draw a plot of two variables with bivariate and univariate graphs.
kinds=scatter,hex,reg,kde

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm",hue='species');
```

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", kind='hex',alpha=.4,size=6);
```

**Pair Plot**

Plot pairwise relationships in a dataset.

```python
sns.pairplot(penguins,hue='sex',diag_kind='hist');
```

```python
sns.pairplot(tips,corner=True);
```

**Grid Plots**

    FacetGrid
    PairGrid

**PairGrid**

```python
g=sns.PairGrid(tips,hue='sex')
g.map(plt.scatter);
```

```python
g=sns.PairGrid(tips)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot);
```

**FacetGrid**

```python
g=sns.FacetGrid(tips,row="time",col="smoker")
g.map(plt.hist,"total_bill");
```
