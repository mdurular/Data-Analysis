# Filling NaN values

```python
df2 = df.fillna(value = 0)
df2
```

```python
df2.isnull().sum().sum()
```

```python
df3 = df.fillna(value = 5)
df3
```

```python
# Filling NaN values with a previous value

df4 = df.fillna(method='pad')
df4
```

```python
df4.isnull().sum()
```

```python
df
```

```python
# Filling NaN values with the next value

df5 = df.fillna(method = 'bfill')
df5
```

```python
df6 = df.fillna(method = 'pad', axis = 1)
df6
```

```python
df7 = df.fillna(method = 'bfill', axis = 1)
df7
```

```python
df
```

```python
# Filling Different NaN Values in Null in different columns

df8 = df.fillna({'society': 'abcd',
                 'balcony':'defg'})
df8
```

```python
# Fillinf NaN value with the mean of a column

df9 = df.fillna(value = df['balcony'].mean())
df9
```
