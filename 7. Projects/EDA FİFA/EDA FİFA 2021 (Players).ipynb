# Average age by country

```python
df_age = df.groupby("Nationality")[["Player_Age"]].mean()
```

```python
df_age.sort_values(by="Player_Age", ascending=False)
```

 Youngest country with average age by country are __Malaysia__ and __Singapore__ , oldest country is  __Korea DPR__



```python

plt.figure(figsize=(24,14))

ax = df.groupby("Player_Age")["Player_Name"].count().plot.bar()
plt.xticks(fontsize = 20,rotation=0)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x()-0.1, p.get_height()+30))
```

# Number of players by country

```python
df_Nationality =df.groupby(["Nationality"])["Player_Id"].count().sort_values(ascending=False).head(10)
```

```python
x = df_Nationality.index
```

```python
y = df_Nationality.values
```

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(x = x, y = y)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x()-0.02, p.get_height()+20)) 
```
