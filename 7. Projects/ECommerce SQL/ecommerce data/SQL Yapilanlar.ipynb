```python
import numpy as np
import pandas as pd
import re
```

# ID DUZENLEME

```python
df = pd.read_csv('prod_dimen.csv')
```

```python
df
```

```python
df.info()
```

```python
def key(x):
    y = re.findall('([0-9]+)', x)
    return int(y[0])

key('Cust_1')
```

```python
df['Prod_id'] = df.Prod_id.apply(key)
```

```python
df
```

```python
df.info()
```

```python
df.to_csv('prod_dimen_new.csv', index=False)
```

# DATE DUZENLEME

```python
df2 = pd.read_csv('orders_dimen.csv')
```

```python
df2
```

```python
df2['Ord_id'] = df2.Ord_id.apply(key)
```

```python
df2
```

```python
df2.info()
```

```python
def date(x):
    y = re.findall('([0-9]+)', x)
    print(y)
    return f'{y[2]}-{y[1]}-{y[0]}'

date('20-10-2010')
```

```python
df2['Order_Date'] = df2.Order_Date.apply(date)
df2
```

```python
df2
```

```python
df2.to_csv('order_dimen_new.csv', index=False)
```
