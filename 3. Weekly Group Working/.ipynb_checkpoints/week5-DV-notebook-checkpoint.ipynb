```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
print(sns.get_dataset_names())
```

```python
mpg = sns.load_dataset("mpg")
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
#describe all features, both numerical and categorical
titanic.describe(include="all")
```

```python
# get number of null values per columns
titanic.isnull().sum()
```

```python
# we can see all missing value in heatmap
sns.heatmap(titanic.isnull(), cbar=True);
```

```python
# we can see numeric column's corr in heatmap
sns.heatmap(titanic.corr(),annot=True, cbar=True);
```

```python
titanic
```

```python
# filter data of columns "sex","pclass","age","fare","survived"
titanic_new =titanic[["sex","pclass","age","fare","survived"]]
titanic_new
```

```python
# draw graph below
g = sns.PairGrid(titanic_new)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# Change default palette to GREEN and RED
g = sns.PairGrid(titanic_new, hue="sex",diag_sharey=False, palette= ["green", "red"])
g.map_diag(sns.kdeplot, shade=True)
g.map_offdiag(sns.scatterplot)
g.add_legend();
```
